```python
# Se verifica si existen nulos y el tipo de data de cada columna
fun.tipo_dato(df_items)
```

#### Se decide eliminar la columna  'playtime_2weeks' dado que existe otra columna que se llama 'playtime_forever', 'items count' que es el conteo de items por usuario que no es necesaria y la de 'user_url' que tampoco se observa nevesaria

```python
# Se eliminan algunas columnas innecesarias
df_items.drop(columns=['playtime_2weeks','items_count','user_url'],inplace=True)
```

```python
# Se verifican todas las filas que contengan nulos 
nulos = df_items[df_items.isnull().any(axis=1)]
nulos
```

```python
# Se revisan si existen filas duplicadas en el DataFrame 
duplicados = df_items.loc[df_items.duplicated()]
duplicados
```

```python
# se eliminan los 59117 datos duplicados
df_items = df_items.drop_duplicates(keep='first') # se conserva la primera aparición de esa fila y se eliminan las apariciones duplicadas subsiguientes.
```

```python
df_items
```

```python
# Cambiar nombre a la columna 'item_id' 
df_items.rename(columns={'item_id' : 'id_game'}, inplace=True)
```

```python
# Ordenamos los id's en forma ascendente y hacemos un reset al index 
df_items=df_items.sort_values(by='id_game', ascending=True)
df_items.reset_index(drop=True, inplace=True)
```

```python
df_items
```

```python
# # Capitalizar la primera letra de cada palabra en el nombre de las columnas
def capitalizar(texto):
    return texto.capitalize()
df_items.columns = df_items.columns.map(capitalizar)
```

```python
# Exportamos a csv
df_items.to_csv('items.csv', index=False)
```
